# Load Data

```python
# Use product and sales related columns
df_product_sales = df[["Product ID",	"Category",	"Sub-Category",	"Product Name","Sales"]]
```

```python
df_product_sales.head()
```

# Aggregate and grouping dataframe

Aggregate and grouping by one column

```python
#Group by 'Category'
grouped = df_product_sales[["Category","Sales"]].groupby('Category')

#Apply an aggregation function (e.g., sum)
aggregated = grouped['Sales'].sum()

#(Optional): Reset the index to make it a DataFrame
aggregated_df = aggregated.reset_index()
```

```python
#The Result
aggregated_df
```

Aggregate and grouping by two columns

```python
#Group by 'Category' and 'Sub-Category'
grouped_2 = df_product_sales[["Category","Sub-Category","Sales"]]\
            .groupby(['Category', 'Sub-Category'])

#Apply an aggregation function (e.g., sum)
aggregated_2 = grouped_2['Sales'].sum()

#(Optional): Reset the index to make it a DataFrame
aggregated_df2 = aggregated_2.reset_index()
```
