```python
# Load the Excel file into a DataFrame
cars = pd.read_excel('cars.xlsx')

# Display the first few rows of the dataset
cars.head()
```

```python
# Display data type information and non-null counts for each column
cars.info()
```

**Find and Drop Duplicate Elements​**

```python
# Identify rows that are duplicates
duplicates = cars[cars.duplicated()]
print(duplicates)
```

```python
# Check for duplicate rows in the entire dataset (returns boolean for each row)
cars.duplicated()
```

```python
# Count the total number of duplicate rows in the dataset
cars.duplicated().sum()
```

```python
# Drop duplicate rows based on the 'Opportunity Number' column;
# this ensures that each Opportunity Number appears only once
cars.drop_duplicates(subset=['Opportunity Number'])
```

**Remove Duplicate Rows in The Dataset​**

```python
# Remove any duplicate rows across the entire dataset and store the result in 'cleaned_data'
cleaned_data = cars.drop_duplicates()

# Display the cleaned dataset
print(cleaned_data)
```

```python
# Confirm no duplicates remain in the cleaned dataset
cleaned_data.duplicated().sum()
```

```python
# Display the structure and summary info of the cleaned dataset
cleaned_data.info()
```

**Extract Unique Elements​**

```python
# Extract and display unique values from the 'Opportunity Number' column
unique_number = cars['Opportunity Number'].unique()
print(unique_number)
```
